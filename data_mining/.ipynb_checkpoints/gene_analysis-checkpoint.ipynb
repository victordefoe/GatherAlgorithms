#### 基于k折验证的xgboost检验

在训练集中，对半划分训练结果精度 以及 5折验证得到的平均精度结果

最小最大归一化对结果未见显著影响,基于树的方法基本使用的信息增益作为决策基础，归一化不影响该类分类器结果

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn import metrics
from  sklearn.datasets  import  make_hastie_10_2
from xgboost.sklearn import XGBClassifier



from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold,cross_val_score

def xgb_kfold(X, y):
    
    import warnings
    warnings.filterwarnings('ignore')

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=2)##test_size测试集合所占比例
    clf = XGBClassifier(

    silent=0 ,#设置成1则没有运行信息输出，最好是设置为0.是否在运行升级时打印消息。
    #nthread=4,# cpu 线程数 默认最大
    learning_rate= 0.2, # 如同学习率
    min_child_weight=1, 
    # 这个参数默认是 1，是每个叶子里面 h 的和至少是多少，对正负样本不均衡时的 0-1 分类而言
    #，假设 h 在 0.01 附近，min_child_weight 为 1 意味着叶子节点中最少需要包含 100 个样本。
    #这个参数非常影响结果，控制叶子节点中二阶导的和的最小值，该参数值越小，越容易 overfitting。
    max_depth=6, # 构建树的深度，越大越容易过拟合
    gamma=0,  # 树的叶子节点上作进一步分区所需的最小损失减少,越大越保守，一般0.1、0.2这样子。
    subsample=1, # 随机采样训练样本 训练实例的子采样比
    max_delta_step=0,#最大增量步长，我们允许每个树的权重估计。
    colsample_bytree=1, # 生成树时进行的列采样 
    reg_lambda=1,  # 控制模型复杂度的权重值的L2正则化项参数，参数越大，模型越不容易过拟合。
    #reg_alpha=0, # L1 正则项参数
    #scale_pos_weight=1, #如果取值大于0的话，在类别样本不平衡的情况下有助于快速收敛。平衡正负权重
    #objective= 'multi:softmax', #多分类的问题 指定学习任务和相应的学习目标
    #num_class=10, # 类别数，多分类与 multisoftmax 并用
    n_estimators=100, #树的个数
    seed=1000 #随机种子
    #eval_metric= 'auc'
     )

    clf.fit(X_train,y_train,eval_metric='auc')

    y_true, y_pred = y_test, clf.predict(X_test)
    print("Accuracy : %.4g" % metrics.accuracy_score(y_true, y_pred))

    # N fold evaluation 
    print('-------N fold evaluation ---------')
    # Kf = KFold(n_splits=3, shuffle=True)
    strKFold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
    scores = cross_val_score(clf, train, label, cv=strKFold)

    print("straitified cross validation scores:{}".format(scores))
    print("Mean score of straitified cross validation:{:.2f}".format(scores.mean()))
    
    return clf
    
    
print('\n\n XgBoost with chi_square selection result')
clf_chi = xgb_kfold(chi_train, label)

print('\n\n XgBoost without chi_square selection result')
clf = xgb_kfold(train, label)
```

这是一个信息增益法筛选特征方法，按理说和决策树内部使用方法一致，此处人工再度实现，便于明晰实验   
对于样本集合D来说，A是D中的某个特征，如下

$$ H_A(D) = -\sum_{i=1}^{n}\frac{|D_i|}{D} log_2 \frac{|D_i|}{D} \tag{1} $$


其中的$ H_A(D) $，是对于样本集合D，将当前特征A作为随机变量（取遍特征A的各个值），求得的经验熵

计算各个特征的条件熵


```python

class IG():
    
    def __init__(self, X, y):
        X = np.array(X)
        n_feature = np.shape(X)[1]  # 烈数
        n_y = len(y) # 样本数
        y

        orig_H = 0
        for i in set(y):  # 计算训练样本对于分类结果的香农熵，
            orig_H += -(y.count(i)/n_y)*math.log2(y.count(i)/n_y)  # 公式1 的实现  

        condi_H_list = []
        for i in range(n_feature):
            feature = X[:,i]  # 抽出一条特征
            sourted_feature = sorted(feature) # 从小到大排序
            
            #量化分级，求出每两个相邻元素之间的平均值作为这两个元素的分割线值
            threshold = [(sourted_feature[inde-1]+sourted_feature[inde])/2 for inde in range(len(feature)) if inde != 0 ]
            # 去除单一极端值的样本
            if max(feature) in threshold:
                threshold.remove(max(feature))
            if min(feature) in threshold:
                threshold.remove(min(feature))
            
            pre_H = 0
            for thre in set(threshold): # 对于每一个独特的分割线阈值
                lower = [y[s] for s in range(len(feature)) if feature[s] < thre] # 从原特征中抽出以这个阈值为分割线的小端
                highter = [y[s] for s in range(len(feature)) if feature[s] > thre] # 从原特征中抽出以这个阈值为分割线的大端
                H_l = 0
                for l in set(lower):
                    H_l += -(lower.count(l) / len(lower))*math.log2(lower.count(l) / len(lower))
                H_h = 0
                for h in set(highter):
                    H_h += -(highter.count(h) / len(highter))*math.log2(highter.count(h) / len(highter))
                temp_condi_H = len(lower)/n_y *H_l+ len(highter)/n_y * H_h
                condi_H = orig_H - temp_condi_H
                pre_H = max(pre_H,condi_H)
            condi_H_list.append(pre_H)

        self.IG = condi_H_list


    def getIG(self):
        return self.IG
```

```python

class IG_D():
    #离散值的计算
    
    def __init__(self, X, y):
        X = np.array(X)
        n_feature = np.shape(X)[1]  # 列数
        n_y = len(y) # 样本数
        
        orig_H = self.calc_ent(y)
        
        condi_H_list = []
        for i in range(n_feature):
            feature = X[:, i]
            condi_H = self.calc_condition_ent(feature, y)
            pre_H = orig_H - condi_H
            condi_H_list.append(pre_H)
        
        self.IG = condi_H_list
            
            
    def getIG(self):
        return self.IG
    
    def calc_ent(self, x):
        """
            calculate shanno ent of x
        """
        x_value_list = set([x[i] for i in range(x.shape[0])])
        ent = 0.0
        for x_value in x_value_list:
            p = float(x[x == x_value].shape[0]) / x.shape[0]
            logp = np.log(p)
            ent -= p * logp

        return ent

    def calc_condition_ent(self, x, y):
        """
            calculate ent H(y|x)
        """

        # calc ent(y|x)
        x_value_list = set([x[i] for i in range(x.shape[0])])
        ent = 0.0
        for x_value in x_value_list:
            sub_y = y[x == x_value]
            temp_ent = self.calc_ent(sub_y)
            ent += (float(sub_y.shape[0]) / y.shape[0]) * temp_ent

        return ent

    def calc_ent_grap(self, x, y):
        """
            calculate ent grap
        """

        base_ent = self.calc_ent(y)
        condition_ent = self.calc_condition_ent(x, y)
        ent_grap = base_ent - condition_ent

        return ent_grap
```

```python

list(label)
s1 = IG(np.array(train), list(label)).getIG()
s2 = IG_D(train, np.array(label)).getIG()
IGS = pd.DataFrame([s1,s2]).transpose()
name = train.columns
IGS['gene_name'] = pd.Series(name)

IGS


```
